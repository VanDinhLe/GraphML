# data prep

```python
visit_nodes = {}
merged = pd.merge(condition, drug, on="visit_occurrence_id", how="inner")
```

```python
merged.head()
```

```python
merged['visit_occurrence_id'].duplicated()
```

```python
from torch_geometric.data import HeteroData
data = HeteroData()
```

```python
# Build mapping from original IDs → numeric indices
visit_map = {vid: i for i, vid in enumerate(merged["visit_occurrence_id"].unique())}
person_map = {pid: i for i, pid in enumerate(merged["person_id_x"].unique())}
disease_map = {did: i for i, did in enumerate(merged["condition_concept_id"].unique())}
drug_map = {drid: i for i, drid in enumerate(merged["drug_concept_id"].unique())}
```

```python
data['visit'].num_nodes = len(visit_map)
data['person'].num_nodes = len(person_map)
data['disease'].num_nodes = len(disease_map)
data['drug'].num_nodes = len(drug_map)
```

```python
import torch

# visit -> person
edge_index_visit_person = torch.tensor([
    merged["visit_occurrence_id"].map(visit_map).values,
    merged["person_id_x"].map(person_map).values
], dtype=torch.long)

# visit -> drug
edge_index_visit_drug = torch.tensor([
    merged["visit_occurrence_id"].map(visit_map).values,
    merged["drug_concept_id"].map(drug_map).values
], dtype=torch.long)

# visit -> condition
edge_index_visit_condition = torch.tensor([
    merged["visit_occurrence_id"].map(visit_map).values,
    merged["condition_concept_id"].map(disease_map).values
], dtype=torch.long)
```

```python
# --- Edges (directed) ---
data["visit", "has_person", "person"].edge_index = edge_index_visit_person
data["visit", "has_drug", "drug"].edge_index = edge_index_visit_drug
data["visit", "has_condition", "disease"].edge_index = edge_index_visit_condition
```

```python
from torch_geometric.transforms import ToUndirected
data = ToUndirected()(data)
```

# draw graph
Use the torch_geometric.utils.to_networkx function to convert your PyG Data object into a networkx graph object.
This makes it possible to use the extensive plotting capabilities of the networkx and matplotlib libraries. 

```python
import networkx as nx
from torch_geometric.utils import to_networkx

# Convert the PyG Data object to a networkx graph
G = to_networkx(data)
```

```python
import matplotlib.pyplot as plt

# Draw the graph
nx.draw(G, with_labels=True, node_color='skyblue', edge_color='gray')

# Display the plot
plt.show()
```

```python
pos = nx.spring_layout(G, k=2.0, scale=5.0, iterations=100)
nx.draw(G, pos, with_labels=False, node_size=50, node_color='orange')

# Display the plot
plt.show()
```
